```python
def extract_elements():

    # Path to your SQLite database
    db_path = './ipl_database.db'

    # Connect to the SQLite database
    conn = sqlite3.connect(db_path)

    # Create a cursor object using the cursor() method
    cursor = conn.cursor()

    # SQL query to fetch unique cities
    unique_cities_query = "SELECT DISTINCT City FROM ipl_match_list;"

    #SQL query to fetch unique venues
    unique_venues_query = "SELECT DISTINCT Venue FROM ipl_match_list;"

    # SQL query to fetch unique Umpire1
    unique_umpire_query = """
    SELECT DISTINCT Umpire1 FROM ipl_match_list
    UNION
    SELECT DISTINCT Umpire2 FROM ipl_match_list;
    """

    #SQL query to fetch unique player names
    # unique_player_query = "SELECT DISTINCT Team1Players FROM ipl_match_list;"

    unique_player_query = """
    SELECT DISTINCT Team1Players FROM ipl_match_list
    UNION
    SELECT DISTINCT Team1Players FROM ipl_match_list;
    """

    #SQL query to fetch unique team names

    unique_team_name_query = """
    SELECT DISTINCT Team1 FROM ipl_match_list
    UNION
    SELECT DISTINCT Team2 FROM ipl_match_list;
    """

    # Execute the queries and fetch the results
    try:
        cursor.execute(unique_cities_query)
        unique_cities = cursor.fetchall()
        unique_cities = [item[0] for item in unique_cities if item[0] is not None]

        cursor.execute(unique_venues_query)
        unique_venues = cursor.fetchall()
        unique_venues = [item[0] for item in unique_venues if item[0] is not None]
        
        cursor.execute(unique_umpire_query)
        unique_umpire = cursor.fetchall()
        unique_umpire = [item[0] for item in unique_umpire if item[0] is not None]

        cursor.execute(unique_team_name_query)
        unique_team_name = cursor.fetchall()
        unique_team_name = [item[0] for item in unique_team_name if item[0] is not None]

        cursor.execute(unique_player_query)
        unique_player = cursor.fetchall()
        unique_player = [item[0] for item in unique_player if item[0] is not None]
        unique_player_name = []
        for i in unique_player:

            # Remove brackets and split the string
            elements = i.strip("[]").split(", ")

            # Trim each element and remove quotes
            elements = [element.strip("'") for element in elements]

            for j in elements:
                unique_player_name.append(j)
        
        unique_player_name = list(set(unique_player_name))

        search_dict = {}
        search_dict['Players'] = unique_player_name
        search_dict['Venues'] = unique_venues
        search_dict['Cities'] = unique_cities
        search_dict['Teams'] = unique_team_name
        search_dict['Umpires'] = unique_umpire
            
        return search_dict
        

    except sqlite3.Error as error:
        print("Error while executing SQL", error)

    finally:
        # Close the database connection
        conn.close()
```

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('ipl_database.db')
cursor = conn.cursor()

# Task 1: Calculate the sum of total_run column elements and the number of instances of isWicketDelivery==1 for innings==y
venue_value = 'Narendra Modi Stadium'  # Replace 'x' with the desired venue value
innings_value = 1  # Replace 1 with the desired innings value
cursor.execute("""
    SELECT
        ipl_match_list.ID,
        SUM(ipl_ball_by_ball.total_run) AS total_runs,
        SUM(CASE WHEN ipl_ball_by_ball.isWicketDelivery = 1 THEN 1 ELSE 0 END) AS wicket_deliveries
    FROM
        ipl_match_list
    JOIN
        ipl_ball_by_ball ON ipl_match_list.ID = ipl_ball_by_ball.ID
    WHERE
        ipl_match_list.Venue = ? AND ipl_ball_by_ball.innings = ?
    GROUP BY
        ipl_match_list.ID
""", (venue_value, innings_value))
task1_results = cursor.fetchall()
print("Task 1 Results:")
for row in task1_results:
    print(row)

# Task 2
# Execute the SQL query to get the maximum over bowled
cursor.execute("""
    SELECT
        MAX(overs) AS max_over_bowled
    FROM
        ipl_ball_by_ball
    JOIN
        ipl_match_list ON ipl_ball_by_ball.ID = ipl_match_list.ID
    WHERE
        ipl_match_list.Venue = ? AND ipl_ball_by_ball.innings = ?
""", (venue_value, innings_value))
max_over_result = cursor.fetchone()

# Print the result
if max_over_result[0] is not None:
    print(f"The maximum over bowled in innings {innings_value} and venue '{venue_value}' is: {max_over_result[0]}")
else:
    print(f"No data found for innings {innings_value} and venue '{venue_value}'")

# Close the connection
conn.close()
```

```python
import sqlite3

def get_max_over_bowled_by_venue_and_innings(venue_value, innings_value):
    conn = sqlite3.connect('ipl_database.db')
    cursor = conn.cursor()

    # Execute the SQL query to get the maximum over bowled for each ID
    cursor.execute("""
        SELECT
            ipl_match_list.ID,
            MAX(ipl_ball_by_ball.overs) AS max_over_bowled
        FROM
            ipl_ball_by_ball
        JOIN
            ipl_match_list ON ipl_ball_by_ball.ID = ipl_match_list.ID
        WHERE
            ipl_match_list.Venue = ? AND ipl_ball_by_ball.innings = ?
        GROUP BY
            ipl_match_list.ID
    """, (venue_value, innings_value))
    max_over_results = cursor.fetchall()

    # Print the results
    # for result in max_over_results:
    #     print(f"ID: {result[0]}, Maximum Overs Bowled: {result[1]}")

    # Close the connection

    conn.close()
    return max_over_results

# Example usage with variable values for venue and innings
venue_value = 'Narendra Modi Stadium'  # Example venue
innings_value = 1  # Example innings
get_max_over_bowled_by_venue_and_innings(venue_value, innings_value)
```

```python
def innings_table_row_creator(venue_value, innings_value):
    conn = sqlite3.connect('./ipl_database.db')
    cursor = conn.cursor()
    cursor.execute("""
        SELECT
            ipl_match_list.ID,
            SUM(ipl_ball_by_ball.total_run) AS total_runs,
            SUM(CASE WHEN ipl_ball_by_ball.isWicketDelivery = 1 THEN 1 ELSE 0 END) AS wicket_deliveries
        FROM
            ipl_match_list
        JOIN
            ipl_ball_by_ball ON ipl_match_list.ID = ipl_ball_by_ball.ID
        WHERE
            ipl_match_list.Venue = ? AND ipl_ball_by_ball.innings = ?
        GROUP BY
            ipl_match_list.ID
    """, (venue_value, innings_value))
    task1_results = cursor.fetchall()

    
    task2_results = get_max_over_bowled_by_venue_and_innings(venue_value, innings_value)


    conn.close()
    return task1_results,task2_results
```

```python
innings_value = 1
id_list = []
batting_lst = []
wickets_lst = []
batting_total_lst = []
wickets_total_lst = []
avg_runs_lst = []
avg_wickets_lst = []
total_avg_lst = []
overs_list = []
avg_overs_list = []
for i in venue_list:
    t1, t2 = innings_table_row_creator(i, innings_value)
    id = ""
    r = ""
    w = ""
    tr = 0
    tw = 0
    for i in t1:
        id+=str(i[0]) + ","
        r+=str(i[1]) + ","
        tr+=i[1]
        w+=str(i[2]) + ","
        tw+=i[2]

    o = ""
    ov = 0
    for i in t2:
        o+=str(i[1]) + ","
        ov+=i[1]

    id_list.append(id)
    batting_lst.append(r)
    wickets_lst.append(w)
    batting_total_lst.append(tr)
    wickets_total_lst.append(tw)

    if len(t1)!=0:
        avg_runs_lst.append(tr/len(t1))
        avg_wickets_lst.append(tw/len(t1))
    else:
        avg_runs_lst.append(0)
        avg_wickets_lst.append(0)
    
    if tw != 0:
        total_avg_lst.append(tr/tw)
    else:
        total_avg_lst.append(0)
    
    overs_list.append(o)

    if len(t2) !=0:
        avg_overs_list.append(ov/len(t2))
    else:
        avg_overs_list.append(0)


venue_innings_1_dict = {}

venue_innings_1_dict["Venue"] = venue_list
venue_innings_1_dict["ID_List"] = id_list
venue_innings_1_dict["Runs_List"] = batting_lst
venue_innings_1_dict["Wickets_List"] = wickets_lst
venue_innings_1_dict["Total_Runs"] = batting_total_lst
venue_innings_1_dict["Total_Wickets"] = wickets_total_lst
venue_innings_1_dict["Average_Runs_Per_Innings"] = avg_runs_lst
venue_innings_1_dict["Average_Wickets_Per_Innings"] = avg_wickets_lst
venue_innings_1_dict["Innings_Average"] = total_avg_lst
venue_innings_1_dict["Overs_List"] = overs_list
venue_innings_1_dict["Average_Overs_Per_Innings"] = avg_overs_list



df_venue_innings_1 = pd.DataFrame(venue_innings_1_dict)

df_venue_innings_1
```
